```python
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from random import random
import requests
from tqdm.notebook import tqdm
from urllib.parse import urljoin
import urllib.parse as urlparse
from urllib.parse import parse_qs
```

```python
base_url='https://www.imdb.com/chart/top/?ref_=nv_mv_250'
response=requests.get(base_url).text
```

```python
soup=BeautifulSoup(response,'html.parser')
```

```python
table=soup.find('table',attrs={'class':'chart full-width'})
```

```python
df=pd.DataFrame(columns=['Rank','Title','IMDB Rating'])
```

```python
for row in table.tbody.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    if(columns!=[]):
        data=columns[1].text.split()
        r=data[0].split(".")
        rank=r[0]
        name=' '.join(data)
        name=name[2:]
        rating=columns[2].text.strip()
        
        df=df.append({'Rank':rank,'Title':name,'IMDB Rating':rating},ignore_index=True)
        
        
```

```python
df['IMDB Rating']=df['IMDB Rating'].replace('\n',' ', regex=True)
```
